```python
import numpy as np
from scipy.fft import fft, fftfreq, ifft
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 1000)
y = -70*x**4 + 120*x**3 - 60*x**2 + 12*x
y_noisy = y + np.random.normal(0, 1, x.shape)
```

```python
plt.plot(x, y_noisy, label="Noisy")
plt.plot(x, y, label="Theory")
plt.legend()
```

```python
freqs = fftfreq(1000, 0.001)
ynf = fft(y_noisy)
plt.plot(freqs, abs(ynf))
plt.xlim(0, max(freqs))
```

```python
ynf[(freqs > 7) | (freqs < -7)] = 0
```

```python
y_processed = ifft(ynf)
plt.plot(x, y_noisy, label="Actual")
plt.plot(x, y_processed, label="Processed")
plt.plot(x, y, label="Actual")
```
